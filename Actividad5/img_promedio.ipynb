```python
import pandas as pd

# Ruta del archivo
ruta_csv = '/content/drive/MyDrive/pruebasNadia/Extraer IMG/dataimg_con_similaridad.csv'
#file_path = '/content/drive/MyDrive/pruebasNadia/Extraer IMG/dataimg_con_similaridad_stopwords.csv'

# Cargar el archivo
df = pd.read_csv(ruta_csv, delimiter=',')
df['texto_figura'] = df['texto_figura'].astype(str)
df['label'] = df['label'].astype(str)
df['tamano'] = pd.to_numeric(df['tamano'], errors='coerce')
df['similaridad'] = pd.to_numeric(df['similaridad'], errors='coerce')

total_df = df.groupby('archivo').agg({
    'texto_figura': ' '.join,
    'label': 'first',
    'tamano': 'sum',
    'similaridad': ['mean', 'max', 'min', 'std', 'median', 'mad']
}).reset_index()

# Calcular la media armónica, coeficiente de variación y curtosis
total_df['media_armonica'] = len(total_df) / sum(1 / total_df['similaridad']['mean'])
total_df['coef_variacion'] = total_df['similaridad']['std'] / total_df['similaridad']['mean'] * 100
total_df['curtosis'] = total_df['similaridad']['mean'] - 3 * total_df['similaridad']['std']

# Renombrar las columnas
total_df.columns = ['archivo', 'texto_completo', 'label', 'tamano_completo', 'promedio_similaridad', 'max_similaridad', 'min_similaridad', 'desviacion_estandar', 'mediana', 'mad', 'media_armonica', 'coef_variacion', 'curtosis']

print(total_df)
# Guardar el resultado en un nuevo archivo CSV
ruta_nuevo_csv = '/content/drive/MyDrive/pruebasNadia/Extraer IMG/dataimg_estadisticas_prieba.csv'
total_df.to_csv(ruta_nuevo_csv, index=False)
```

```python
import pandas as pd

# Ruta del archivo
ruta_csv = '/content/drive/MyDrive/pruebasNadia/Extraer IMG/dataimg_con_similaridad.csv'

# Cargar el archivo
df = pd.read_csv(ruta_csv, delimiter=',')
df['texto_figura'] = df['texto_figura'].astype(str)
df['label'] = df['label'].astype(str)
df['tamano'] = pd.to_numeric(df['tamano'], errors='coerce')
df['similaridad'] = pd.to_numeric(df['similaridad'], errors='coerce')

# Agrupar por el campo 'archivo' y calcular las sumas y promedios
total_df = df.groupby('archivo').agg({
    'texto_figura': ' '.join,
    'label': 'first',
    'tamano': 'sum',
    'similaridad': ['mean', 'min', 'max']
}).reset_index()

# Cambiar el nombre de las columnas
total_df.columns = ['archivo', 'texto_completo', 'label', 'tamano_completo', 'promedio_similaridad', 'similitud_minima', 'similitud_maxima']

# Calcular los valores adicionales
total_df['promedio'] = total_df['promedio_similaridad']['mean']
total_df['similitud_minima'] = total_df['similitud_minima']['min']
total_df['similitud_maxima'] = total_df['similitud_maxima']['max']
total_df['desviacion_estandar'] = df.groupby('archivo')['similaridad'].std()
total_df['media_armonica'] = 2 / (1/total_df['promedio'] + 1/total_df['similitud_maxima'])
total_df['mediana'] = df.groupby('archivo')['similaridad'].median()
total_df['coeficiente_variacion'] = (total_df['desviacion_estandar'] / total_df['promedio']) * 100

# Calcular el coeficiente de correlación de Pearson
coeficientes_pearson = []
for archivo, grupo in df.groupby('archivo'):
    coeficiente = grupo['tamano'].corr(grupo['similaridad'])
    coeficientes_pearson.append(coeficiente)

total_df['coeficiente_pearson'] = coeficientes_pearson

# Guardar el resultado en un nuevo archivo CSV
ruta_nuevo_csv = '/content/drive/MyDrive/pruebasNadia/Extraer IMG/dataimg_promedios_prueba.csv'
total_df.to_csv(ruta_nuevo_csv, index=False)

print(total_df)
```
