```python
import random
import pandas as pd

#####------PARAMETROS DE FILTRADO-----------########
tamano_tweet = 280
ruta_archivo_palabras = "/content/drive/My Drive/n_gramas/PalabrasparaMujerEjemplo.txt"
ruta_archivo_mujer = "/content/drive/My Drive/n_gramas/corpus_Ngramas.txt"
archivo_canciones = "/content/drive/My Drive/n_gramas/es_Canciones2Negativas.csv"
palabras_bolsa_1 = 1
palabras_bolsa_2 = 2
porcentaje = 0.2  # 20%
#####------termina PARAMETROS DE FILTRADO----########

# Leer CSV y cargarlo en un solo dataframe
df = pd.read_csv(archivo_canciones, delimiter=',', encoding='latin1')
# Verificar los primeros registros del DataFrame
print(df.head())
print(df.columns)

# Cargar palabras y verbos desde archivos de texto
with open(ruta_archivo_palabras, "r") as f:
    palabras = set(linea.strip() for linea in f.readlines())

with open(ruta_archivo_mujer, "r") as f:
    misoginas = [linea.strip() for linea in f.readlines() if linea.strip()]

# Generar n-gramas según la etiqueta (1 para positivo, 0 para negativo)
def generar_ngramas_segun_etiqueta(cancion, etiqueta):
    max_length = tamano_tweet
    palabras_cancion = cancion.split()
    ngramas = []
    if etiqueta == 1:
        for n in range(3, len(palabras_cancion) + 1):
            for i in range(len(palabras_cancion) - n + 1):
                ngrama = " ".join(palabras_cancion[i:i + n])
                palabras_ngrama = ngrama.split()
                if len(ngrama) <= max_length and any(palabra in palabras for palabra in palabras_ngrama) and sum(palabras_bolsa_1 for palabra in palabras_ngrama if palabra in misoginas) >= palabras_bolsa_2:
                    ngramas.append(ngrama)
    elif etiqueta == 0:
        for n in range(1, len(palabras_cancion) + 1):
            for i in range(len(palabras_cancion) - n + 1):
                ngrama = " ".join(palabras_cancion[i:i + n])
                if len(ngrama) <= max_length:
                    ngramas.append(ngrama)
        # Revolver el índice de la lista de n-gramas negativos
        random.shuffle(ngramas)
        # Seleccionar el 20% de los n-gramas
        total_ngramas = len(ngramas)
        ngramas = ngramas[:int(total_ngramas * porcentaje)]
    return ngramas

# Aplica las funciones a cada fila del dataframe
ngramas_filtrados = []
for index, row in df.iterrows():
    etiqueta = row['misogynous']
    tweet = row['tweet'].lower()
    ngramas = generar_ngramas_segun_etiqueta(tweet, etiqueta)
    ngramas_filtrados.extend(ngramas)

# Crear un DataFrame con los n-gramas y las etiquetas
etiquetas = [1 if etiqueta == 1 else 0 for _ in ngramas_filtrados]
df_ngramas = pd.DataFrame({'tweet': ngramas_filtrados, 'label': etiquetas})

# Guardar el DataFrame en un archivo CSV
ruta_archivo_ngramas = "/content/drive/My Drive/n_gramas/ngramas_filtrados_prueba5.csv"
df_ngramas.to_csv(ruta_archivo_ngramas, index=False)

print(f"N-gramas guardados en: {ruta_archivo_ngramas}")
```
